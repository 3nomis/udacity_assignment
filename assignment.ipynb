```python
from torchvision import datasets, transforms
import torch
```

```python
import numpy as np
import torch.utils.data.sampler as smp

batch_size = 30

split_rate = 0.2


# In each batch images need to be of the same sizes
trasformer = transforms.Compose([
    transforms.Resize(size=(300,300)),
    transforms.RandomRotation(degrees=20),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5,0.5,0.5],std=[0.5,0.5,0.5])
])

train_set = datasets.ImageFolder(root='flower_data/train/', transform=trasformer)
tot_train = len(train_set)
valid_set = datasets.ImageFolder(root='flower_data/valid/', transform=trasformer)


tot_indexes = list(range(0, tot_train))
np.random.shuffle(tot_indexes)

train_indexes = tot_indexes[round(split_rate*tot_train):]
test_indexes = tot_indexes[0:round(split_rate*tot_train)]

train_sampler = smp.SubsetRandomSampler(indices=train_indexes)
test_sampler = smp.SubsetRandomSampler(indices=test_indexes)


train_loader = torch.utils.data.DataLoader(dataset=train_set, 
                                           batch_size=batch_size, 
                                           sampler=train_sampler)

test_loader = torch.utils.data.DataLoader(dataset=train_set, 
                                           batch_size=batch_size, 
                                           sampler=test_sampler)

valid_loader = torch.utils.data.DataLoader(dataset=valid_set, 
                                           batch_size=batch_size, 
                                           shuffle=True)
```
